```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stocks = pd.read_csv("results.csv")
stocks.head()
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x = "hour", y = "max_high_per_hour", data = stocks, hue = "name").set(ylabel=' Maximum Stock High (US Dollars)')
plt.show()

# Observations:
    # shopify (SHOP) has the highest value for stock high, followed by Netflix(NFLX) and then facebook(FB)
    # Maximum stock high remains constant each hour of November 30th for all companies except shopify; at about 10 AM, it
      # starts to drop, but the maximum stock high value remains above $1400 for shopify
```

```python
tickers = ["FB", "SHOP", "BYND", "NFLX", "PINS", "SQ", "TTD", "OKTA", "SNAP", "DDOG"]
colors = ["g", "b", "r", "y", "k", "orange", "m", "skyblue", "yellow", "pink"]

fig, axes  = plt.subplots(5,2, figsize=(12,20))

for i in range(len(tickers)):
    stocks_com = stocks[stocks['name'] == tickers[i]]
    
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    sns.boxplot(x = "name", y = "max_high_per_hour", ax= ax, data = stocks_com, color = colors[i], 
                boxprops=dict(alpha=.6)).set( xlabel= '', ylabel='Maximum Stock High (US Dollars)')

plt.show()

# Observations:
     # As observed above, the stock prices don't change much throughout the day for most companies-> maximum stock 
      # high values has small variation, as can be seen below. SHOP (shopify) has slighlty high variation
    
```
